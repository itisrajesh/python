### Set

```python
#@title { vertical-output: true}

my_set = {1, 2, 3, 4, 5}  # Creating a set
print("My Set:", my_set)
my_set.add(6)  # Adding elements to a set
print("After adding 6:", my_set)
my_set.update([7, 8, 9])  # Updating a set with multiple values
print("After updating with [7, 8, 9]:", my_set)
my_set.remove(2)  # Removing elements from a set
print("After removing 2:", my_set)
my_set.discard(3)  # Discarding an element from a set (does nothing if element is not present)
print("After discarding 3:", my_set)
copied_set = my_set.copy()  # Copying a set
print("Copied Set:", copied_set)
my_set.clear()  # Clearing a set
print("After clearing:", my_set)

# Set operations
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
print("Set1", set1)
print("Set2", set2)
print("Union:", set1.union(set2))  # Union
print("Intersection:", set1.intersection(set2))  # Intersection
print("Difference:", set1.difference(set2))  # Difference
print("Symmetric Difference:", set1.symmetric_difference(set2))  # Symmetric Difference
print("Is set1 a subset of set2?", set1.issubset(set2))  # Subset
print("Is set1 a superset of set2?", set1.issuperset(set2))  # Superset
print("Is set1 disjoint with set2?", set1.isdisjoint(set2))  # Disjoint (no element is common b/w them)
popped_element = set1.pop()  # Pop
print("Popped element:", popped_element)
print("After popping:", set1)
print("Length of set1:", len(set1))  # Len
print("Is 2 in set1?", 2 in set1)  # In

# Assignment 2 - Common element b/w tuple
r_tuple = input("Enter the comma separated tuple 1: ")
tuple_1 = tuple(int(x) for x in str(r_tuple).split(","))
r_tuple = input("Enter the comma separated tuple 2: ")
tuple_2 = tuple(int(x) for x in str(r_tuple).split(","))
common = set(tuple_1).intersection(set(tuple_2))
print(f"Common between \n{tuple_1} \n{tuple_2} is \n{common}")

# Assignment 3 - Extract unique letter from a given test using set
para = input("Enter the paragraph: ")
unique_letters = set(x for x in str(para).split(" "))
ans = sorted(unique_letters, key=lambda s: s.lower())
ans = " ".join(ans)
print(ans)

# Assignment 4 - Simulate Venn diagram-like operator using in python set
# fruits = {'apple', 'mango', 'banana', 'tomato', 'grapes'}
# vegetables = {'radish', 'eggplant', 'tomato', 'grapes', 'beans'}
# grains = {'rice', 'wheat', 'barley', 'beans', 'grapes'}
fruits = {'a', 'b', 'c', 'd', 'e'}
vegetables = {'a', 'e', 'f', 'g', 'h'}
grains = {'a', 'h', 'i', 'j', 'k'}

print(f"fruits: {fruits}")
print(f"vegetables: {vegetables}")
print(f"grains: {grains}\n")

common_all = fruits.intersection(vegetables).intersection(grains)
not_common_in_any = fruits.symmetric_difference(vegetables).symmetric_difference(grains)
fruits_unique = fruits.difference(vegetables.union(grains))
vegetables_unique = vegetables.difference(fruits_unique.union(grains))
grains_unique = grains.difference(vegetables.union(fruits))

print("Common all: ", common_all)
print("Not Common in any : ", not_common_in_any)
print("fruits unique: ", fruits_unique)
print("vegetables_unique: ", vegetables_unique)
print("grain unique: ", grains_unique)
```

```python
dict1 = {"name": "John", "age": 30, "city": "New York"}
print("DICT1 : ", dict1)
dict5 = {"name": "John Wick", "country": "USA"}
dict1.update(dict5)
print("DICT1: ", dict1)
# print("DICT5 - update : ", dict5)
dict1.clear()
print("DICT1 - clear : ", dict1)
dict2 = {"name": "Jane", "age": 25, "city": "Los Angeles"}
print("DICT2 : ", dict2)
dict3 = dict2.copy()
dict4 = {}.fromkeys(["name", "age", "city"], "default")
print("DICT4 - fromkeys : ", dict4)
print("DICT2 - [] : ", dict2["name"])
print("DICT2 - get : ", dict2.get("name"))
print("DICT2 - items : ", dict2.items())
print("DICT2 - keys : ", dict2.keys())
print("DICT2 - pop : ", dict2.pop("age"))  # del dict2["age"]
print("DICT2 - popitem : ", dict2.popitem())
print("DICT2 - setdefault : ", dict2.setdefault("country", "USA"))
dict2.update({"name": "Jim", "age": 35})
print("DICT2 - update : ", dict2.values())

movies = {'a': 1, 'b': 2, 'c': 3}
print(f"Max of {movies} is {max(movies, key=movies.get)} ")


# Assignment 5 - Dict with int key and cube value
dict_cube = {}
for i in range(1, 6):
    dict_cube[i] = i ** 3
print(dict_cube)

# Assignment 6 - a->e with Ascii
ascii_dict = {}
for letter in range(ord('a'), ord('e') + 1):
    ascii_dict[chr(letter)] = letter
print(ascii_dict)

# Assignment 7 - Inventory system
# import json

item_dict = {}
while True:
    print("Enter choice:")
    choice = input("1. Add item\n2. Update item\n3. Total Inventory value\n4. Display Inventory\n 5.exit")
    match choice:
        case '1':
            item = input("Item Name: ")
            quant = int(input("Item Quantity: "))
            price = float(input("Item Price: "))
            item_dict[item] = {'quantity': quant, 'price': price}
        case '2':
            item = input("Enter item name to update")
            choice = input("Update: 1. Quantity 2. Price")
            match choice:
                case '1':
                    new_val = int(input("Item Quantity: "))
                    item_dict[item].update({'quantity': new_val})
                case '2':
                    new_val = float(input("Item Price: "))
                    item_dict[item].update({'price': new_val})
        case '3':
            total = 0
            for (key, val) in item_dict.items():
                total += val['quantity'] * val['price']
            print(f"Total inventory value: {total}")
        case '4':
            # print(json.dumps(item_dict, indent=4))
            # print(f"{'ITEM': >2}{'Quantity': >2}{'Price': >2}\n--------------------------------------")
            print(f"{'ITEM'}\t\t{'Quantity'}\t\t{'Price'}\n--------------------------------------")
            for key, val in item_dict.items():
                print(f"{key}\t\t{val['quantity']}\t\t{val['price']}\t\t")
        case _:
            break
```

### Comprehension

```python
numbers = [1, 2, 3, 4, 5]
squared_numbers = [n ** 2 for n in numbers]
print(squared_numbers)  # Output: [1, 4, 9, 16, 25]

numbers = [1, 2, 3, 4, 5]
even_numbers = [n for n in numbers if n % 2 == 0]
print(even_numbers)  # Output: [2, 4]

matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flattened_matrix = [num for row in matrix for num in row]
print(flattened_matrix)  # Output: [1, 2, 3, 4, 5, 6, 7, 8, 9]

numbers = [1, 2, 3, 4, 5]
even_numbers_greater_than_2 = [n for n in numbers if n % 2 == 0 and n > 2]
print(even_numbers_greater_than_2)  # Output: [4]

fruits = ['apple', 'banana', 'cherry']
fruit_dict = {fruit: len(fruit) for fruit in fruits}
print(fruit_dict)  # Output: {'apple': 5, 'banana': 6, 'cherry': 6}

# Assignment 11: Extract vowels from string
my_string = "djksj dhfhh reet sfdsd fd."
vowels = [c for c in my_string if c.lower() in "aeiou"]
print(f'={vowels}')

# Assignment 12: Invert dictionary, swap key-values
my_dict = {'a': 1, 'b': 2, 'c': 6}
print("Original:", my_dict)
my_dict = {y: x for x, y in my_dict.items()}
print("Inverted: ", my_dict)



```
