### Assignment 1: Create a Program that Converts a List of Tuples into a Dictionary

#### Input:
- Prompt the user to enter a list of tuples, where each tuple contains two elements (key, value).

#### Logic:
- Convert the list of tuples into a dictionary.
- Ensure that keys in the dictionary are unique. If there are duplicate keys, use the last occurrence.

#### Output:
- Display the resulting dictionary.

```python
data_list = []
while True:
    print("1. Add data. 2. Exit")
    choice = input("Choice : ")
    match choice:
        case '1':
            key = input("Key: ")
            value = input("Value: ")
            data_list.append((key,value))
        case '2':
            break
        case _:
            print("Invalid choice")

data_dict = {}
for tup in data_list:
    data_dict[tup[0]]=tup[1]

print(data_dict)


print({ tup[0]:tup[1] for tup in data_list})
```

### Assignment 2: Write a Program that Counts the Frequency of Each Element in a List and Stores the Results in a Dictionary

#### Input:
- Prompt the user to enter a list of elements (numbers, strings, etc.).

#### Logic:
- Use a loop to count the occurrences of each element in the list.
- Store the element and its frequency as key-value pairs in a dictionary.

#### Output:
- Display the dictionary with the frequency of each element.

```python
str = input("Enter a string : ")
lst = [x for x in str]
dict1 = {x: lst.count(x) for x in lst}
print(dict1)

dict2 = {}
for x in lst:
    if x in dict2:
        dict2[x]+=1
    else:
        dict2[x]=1
print(dict2)
```

### Assignment 3: Create a Country Capitals Quiz Program Using Dictionaries

#### Input:
- Store a dictionary of countries and their capitals.
- Ask the user to match countries with their capitals.

#### Logic:
- Allow the user to take the quiz and provide feedback after each question.
- Ask if the user wishes to play more.
- Calculate the final score.

#### Output:
- Display the correct answer when the user gets it wrong.
- Show the final score versus the number of attempted questions.

```python
import random

country_capitals = {
    'India': 'Delhi',
    'Australia': 'Canberra',
    'Canada': 'Ottawa',
    'France': 'Paris',
    'Germany': 'Berlin',
    'Japan': 'Tokyo',
    'United Kingdom': 'London',
    'United States': 'Washington'
}



total_attempts = 0
total_wins = 0

while True:
    country = random.choice(list(country_capitals.keys()))
    capital = country_capitals[country]

    print(f"Capital of {country}?")
    user_answer = input()

    total_attempts += 1
    if user_answer.lower() == capital.lower():
        print("Correct!")
        total_wins += 1
    else:
        print(f"Incorrect! Answer : {capital}.")

    print(f"Total Attempts: {total_attempts}, Total Wins: {total_wins}")

    continue_quiz = input("\nContinue? (yes/no): ").lower()

    if continue_quiz != 'yes':
        break
```

### Assignment 5: Use Dictionary Comprehensions to Count the Frequency of Elements in a List

#### Task:
- Given a list of words `['apple', 'banana', 'apple', 'orange', 'banana', 'apple']`, use a dictionary comprehension to create a dictionary that counts the frequency of each word in the list.
- Print the resulting dictionary.

```python
# Given list of words
words = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple']

# Use dictionary comprehension to count frequency of each word
word_frequency = {word: words.count(word) for word in set(words)}

# Print the resulting dictionary
print(word_frequency)
```

### Assignment 7: Merge Two Lists into a Dictionary Using Dictionary Comprehensions

#### Input:
- Given two lists:
  - `keys = ['name', 'age', 'city']`
  - `values = ['Alice', 25, 'New York']`
  
- Use a dictionary comprehension to combine them into a dictionary where the elements of `keys` are the keys and the elements of `values` are the corresponding values.

#### Output:
- Print the resulting dictionary.

```python
# Given lists
keys = ['name', 'age', 'city']
values = ['Alice', 25, 'New York']

# Dictionary comprehension
result_dict = {key: value for key, value in zip(keys, values)}

# Print the resulting dictionary
print(result_dict)
```

### Assignment 5: Use find() and replace()

#### Task:
- **Find the First Occurrence of a Substring**: `text.find(substring)`
- **Find All Occurrences of a Substring**: `text.find(substring)`
- **Replace First Occurrence of a Substring**: `text.replace(old, new, 1)`
- **Replace All Occurrences of a Substring**: `text.replace(old, new)`

```python
text = "Hello, welcome to the world welcomeof Python. Welcome again!"
substring = "welcome"

# First Occurrence
first_occurrence = text.find(substring)
print(f"First occurrence of '{substring}': {first_occurrence}")

# All Occurrences
start = 0
while start < len(text):
    start = text.find(substring, start)
    if start == -1:
        break
    print(f"Found '{substring}' at index: {start}")
    start += len(substring)

# Replacing Substrings
old = "welcome"
new = "greetings"

# Replace the First Occurrence
replaced_text_first = text.replace(old, new, 1)
print(f"After replacing first occurrence: {replaced_text_first}")

# Replace All Occurrences
replaced_text_all = text.replace(old, new)
print(f"After replacing all occurrences: {replaced_text_all}")
```

### Assignment 6: Replace Every Alternate Word in a Paragraph

#### Input:
- A single string paragraph that represents a paragraph of text.

#### Output:
- A new string where every alternate word in the input paragraph is replaced with the word 'replaced'.

#### Logic:
1. **Use the split() method** to divide the paragraph into a list of words.
2. **Replacing Alternate Words**:
   - Iterate through the list of words.
   - Use the index of each word to determine if it is an alternate word.
   - Replace every second word (i.e., words at odd indices) with the word 'replaced'.
3. **Joining Words**:
   - Combine the list of modified words back into a single string using the join() method, ensuring that words are separated by spaces.

```python
str='Lorem Ipsum is simply dummy text of the printing and typesetting industry. Lorem Ipsum has been the industry standard dummy text ever since the 1500s, when an unknown printer took a galley of type and scrambled it to make a type specimen book. It has survived not only five centuries, but also the leap into electronic typesetting, remaining essentially unchanged.'
words = str.split()
for i in range(1, len(words), 2):
    words[i] = 'replaced'
print(' '.join(words))
```
